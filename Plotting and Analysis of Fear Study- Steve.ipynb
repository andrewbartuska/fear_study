#### Initial Attempts to Organize and Automate Plotting Trials

```python
dirname = "data"
subj1 = "steve/faces"
subj2 = "shariq/faces"
```

```python
def run_trials(subject, first_trial, last_trial, norm_method):    
    subj = subject + "/faces"
    
    for run_number in range(first_trial, last_trial+1):
        numb = str(run_number)
        a = "run" + numb + ".csv"
        b = "run" + numb + "_sync_pulses.json"
        c = "faces_run" + numb + ".json"

        norm_method(os.path.join(dirname, subj, a), os.path.join(dirname, subj, b), os.path.join(dirname, subj, c))
```

### Modified Code for Readibility and Ease of Running Individual Trials
New function created in fear_study.py, plot_trial with arguments dirname, subject, trial_number and norm_method

- Dirname = directory where data is stored
- subject = subject name on trial
- trial_number
- norm_method = Normalization method (*Currently "sub" for subtractive and "div" for division*)

#### Strengths

- Allows easy comparison of individual runs with each other
- Easy duplication on separate subjects
- Easy to add normalization methods in fear_study.py file for further/changed comparisons

#### Limitations

- Dependent on naming system in directory (where all files are named based on subject and trial_number without additional characters)
- Plotting of all trials by a single subject requires individual function calls
