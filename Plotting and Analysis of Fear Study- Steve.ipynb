```python
dirname = "data"
subj1 = "steve/faces"
subj2 = "shariq/faces"
```

```python
def run_trials(subject, first_trial, last_trial, RENAME):    
    subj = subject + "/faces"
    
    for run_number in range(first_trial, last_trial+1):
        numb = str(run_number)
        a = "run" + numb + ".csv"
        b = "run" + numb + "_sync_pulses.json"
        c = "faces_run" + numb + ".json"

        RENAME(os.path.join(dirname, subj, a), os.path.join(dirname, subj, b), os.path.join(dirname, subj, c))
```

```python
print "Subtraction"
run_trials("steve", 1, 5, fs.run_all_sub)
print "Division"
run_trials("steve", 1, 5, fs. run_all_div)
```

```python
print "Subtraction"
run_trials("shariq", 4, 4, fs.run_all_sub)
print "Division"
run_trials("shariq", 4, 4, fs.run_all_div)
```
